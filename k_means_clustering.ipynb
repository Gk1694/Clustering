# K-Means Clustering

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3,4]].values
```

## Using the elbow method to find the optimal number of clusters

```
from sklearn.cluster import KMeans
wcs = []
for i in range(1,11):
  kmeans = KMeans(n_clusters= i , init="k-means++",random_state= 42)
  kmeans.fit(X)
  wcs.append(kmeans.inertia_)
plt.plot(range(1,11),wcs)
plt.title("the elbow method")
plt.xlabel('number of cluster')
plt.ylabel('wcss')
plt.show()
```

## Training the K-Means model on the dataset

```
kmeans = KMeans(n_clusters= 5 , init='k-means++', random_state= 42)
y_predict = kmeans.fit_predict(X)
```

```
print(y_predict)
```

## Visualising the clusters

```
cl = ['red','blue','green','cyan','magenta']
for i in range(5):
  plt.scatter(X[y_predict == i,0],X[y_predict == i, 1], s= 100, c = cl[i], label = 'cluster '+ str(i+1))
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 300 , c = 'y', label = 'centriods')
plt.title('Cluster of customer')
plt.xlabel('Anual income')
plt.ylabel('Spending Score from 1 to 100')
plt.legend()
plt.show()
```
